```python
import numpy as np
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
```

```python
x = np.array([6, 3, 5, 2, 1, 4, 9, 7, 8])
y = np.array([2, 1, 3 ,5 ,3 ,9 ,8, 10, 7])
```

```python
# sort the data in x and rearrange y accordingly
sortId = np.argsort(x)
x = x[sortId]
y = y[sortId]
```

```python
maximums = argrelextrema(y, np.greater)
print(maximums)
```

```python
minimum = argrelextrema(y, np.less)
print(minimum)
```

```python
# this way the x-axis corresponds to the index of x
maximum = argrelextrema(y, np.greater)  # (array([1, 3, 6]),)
minimum = argrelextrema(y, np.less)  # (array([2, 5, 7]),)
plt.plot(x-1, y)
plt.show()
```
